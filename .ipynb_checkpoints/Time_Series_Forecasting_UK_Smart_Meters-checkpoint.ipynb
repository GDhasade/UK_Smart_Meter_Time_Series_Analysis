```python
def PlotCorrelationDualCol(dataframe, col1, col2):
    fig, ax1 = plt.subplots(figsize = (25,10))
    ax1.plot(dataframe.day, dataframe[col1], color = 'tab:orange', label=col1)
    ax1.plot(dataframe.day, dataframe[col2], color = 'tab:pink', label=col2)
    ax1.set_ylabel('Temperature')
    ax1.legend()
    ax2 = ax1.twinx()
    ax2.plot(dataframe.day,dataframe.avg_energy,color = 'tab:blue')
    ax2.set_ylabel('Average Energy/Household',color = 'tab:blue')
    #ax2.legend(bbox_to_anchor=(0.0, 1.02, 1.0, 0.102))
    plt.title('Energy Consumption and Temperature')
    fig.tight_layout()
    plt.show()
    

def PlotCorrelationSingleCol(dataframe,col1):
    fig, ax1 = plt.subplots(figsize = (25,10))
    ax1.plot(dataframe.day, dataframe[col1], color = 'tab:orange', label=col1)
    ax1.set_ylabel('Humidity',color = 'tab:orange')
    
    ax2 = ax1.twinx()
    ax2.plot(dataframe.day,dataframe.avg_energy,color = 'tab:blue',label="Avg Energy")

    ax2.set_ylabel('Average Energy/Household',color = 'tab:blue')
    
    
    plt.title('Energy Consumption and '+col1)
    fig.tight_layout()
    plt.show()
```

## Clustering
- To perform clustering - basic requirement is data must be normally distributed / scalar.
- To make data scalar here i am using - MinMaxScalar technique.

```python
Weather_Data = TotalElecUse_Weather[['temperatureMax','humidity','windSpeed']]
Weather_Data
```

### Observation:
- As data in temperatureMax having 2 digit value, humidity in between 0-1 also windspeed in some other range.
- For clustering data we have to perform scaling this data.

```python
from sklearn.preprocessing import MinMaxScaler

ScalerModel = MinMaxScaler()
```

```python
WeatherScaledData = ScalerModel.fit_transform(Weather_Data)
```

```python
WeatherScaledData
```

- Now data in each column in range between 0-1.
- Now clustering can be done.
- But next question is - what will be the k value?
    - To decide k value- We use elbow method.

```python
# Find out optimal K value
from sklearn.cluster import KMeans
ConsiderKValues = range(1, 20)

kmeans = [KMeans(n_clusters=i) for i in ConsiderKValues]
kmeans
```

##### We have 0-19 i.e 20 clusters
- To perform get optimal value of k.
- We plots SSE (Sum Squared Error value on plot and find out elbow value.

```python
# Find out SSE i.e Sum Squared Error for each cluster
SSE = [kmeans[i].fit(WeatherScaledData).score(WeatherScaledData) for i in range(len(kmeans))]
SSE
```

```python
plt.plot(ConsiderKValues,SSE)
plt.xlabel('No. of Clusters')
plt.ylabel('SSE score')
plt.title('Elbow_Curve')
plt.show()
```

### Observation:
- As elbow we get between 0-5.
- Exact elbow at clust 3.
- Hence out optimal k value is 3.

```python
# Weather data cluster
kmeans = KMeans(n_clusters=3, max_iter=600, algorithm = 'auto')
kmeans.fit(WeatherScaledData)
```

```python
# add the cluster value - inshort the weather condition for particular day into original dataframe.
TotalElecUse_Weather['Weather_Condition'] = kmeans.labels_
```

```python
TotalElecUse_Weather['Weather_Condition'].unique()
```

- 0,1,2 indicate cluster 1, 2 and 3 respectively.


```python
# Let plot data on scatter plot and see the distribution on data based on clusters

fig, ax1 = plt.subplots(figsize = (10,7))
ax1.scatter(TotalElecUse_Weather.temperatureMax, 
            TotalElecUse_Weather.humidity, 
            s = TotalElecUse_Weather.windSpeed*10,
            c = TotalElecUse_Weather.Weather_Condition)
ax1.set_xlabel('Temperature')
ax1.set_ylabel('Humidity')
plt.show()
```

- Each colour point indicate the weather condition for particular day.

### Validate data stationary or not
- First we check our data is stationary or not.
- we perform hypothesis test - Dickey-Fuller test.

```python
from statsmodels.tsa.stattools import adfuller
# Hypothesis test
# H0 - It is not stationary
# H1 - It is stationary

def adfuller_test(dataframe):
    result = adfuller(dataframe, autolag='AIC') # result give 5 different values as defined lables below
    
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    
    # printing lables and result based on p value.
    
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adfuller_test(Model_Data.avg_energy)
```

## observation
- Data is not stationary to perform time series analysis need to make it stationary.
- To make it stationary taking differencing of avg_energy by shifting data by 12 days.

```python
Model_Data['Seasonal_Difference']=Model_Data['avg_energy']-Model_Data['avg_energy'].shift(12)
```

```python
## Again test dickey fuller test
adfuller_test(Model_Data['Seasonal_Difference'].dropna())
```

## Result: 
- Difference data is stationary

```python
Model_Data['Seasonal_Difference'].plot(figsize=(25,10))
```

- As now our data is stationary.

### Perform ARIMA seasonal forecasting
- Basically Arima take 3 values
    - p : AR i.e auto regression model lags
    - d : Difference
    - q : Moving average lags
    
- To get p and q value we find auto correlation and partial auto correlation.

```python
# import auto correlation and partial auto corelation plots from statsmodel library 
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(Model_Data['Seasonal_Difference'].dropna(),lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(Model_Data['Seasonal_Difference'].dropna(),lags=40,ax=ax2)
```

# Observation:
- We find the AR Value i.e p : using partial auto correlation plot
    - In graph at 1 it get sudden shutsoff and go down.
    - So p i.e AR model lag value is 1
- Q value we check auto correlation graph - Usally the graph have exponential shut down.
    - As in auto correlation graph - exponential shutdown observed for some value and then data juggle up and down.
    - Here, we take Q is moving average lags - value is 4.
    
- D value
    - We shift data i.e seasonal difference only Once.
    - Hence D i.e difference value is 1
    
- Summary:
    - P = 1
    - Q = 4
    - D = 1
- (p,d,q)
- Seasonal_order (1,1,4,12)

```python
# import required libraries for seasonal forecasting
import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=sm.tsa.statespace.SARIMAX(Model_Data['avg_energy'],order=(1, 1, 1),seasonal_order=(1,1,1,12))
results=model.fit()
```

```python
# now we try to forcast the last few index avg_energy usage 
Model_Data['forecast']=results.predict(start=750,end=827,dynamic=True)
Model_Data[['avg_energy','forecast']].plot(figsize=(20,10))
```
